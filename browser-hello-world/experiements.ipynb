```python
from langchain_google_genai import ChatGoogleGenerativeAI
from browser_use import Agent
import asyncio
import nest_asyncio
from dotenv import load_dotenv

load_dotenv()
nest_asyncio.apply()  # Allow nested async operations in Jupyter

async def main():
    agent = Agent(
        task="Find the top (singular) restaurant with the best reviews in New York City",
        llm=ChatGoogleGenerativeAI(model="gemini-2.0-flash-lite-preview-02-05"),
    )
    
    history = await agent.run()

    try:
        actions = history.model_actions()
        print("Actions: ", actions)
        
        done_result = next((action['done']['text'] for action in actions if 'done' in action), None)
        print(done_result)
    except AttributeError:
        print("Error: `model_actions()` is not available. Check `history` output.")

await main()
```

```python
from browser_use import Agent, Browser, BrowserConfig
from langchain_google_genai import ChatGoogleGenerativeAI
import asyncio
# Configure the browser to connect to your Chrome instance
browser = Browser(
    config=BrowserConfig(
        # Specify the path to your Chrome executable
        chrome_instance_path='/usr/bin/google-chrome',  # macOS path
        # For Windows, typically: 'C:\\Program Files\\Google\\Chrome\\Application\\chrome.exe'
        # For Linux, typically: '/usr/bin/google-chrome'
    )
)

# Create the agent with your configured browser
agent = Agent(
    task="Find me one of those funny sad AI generatd cat videos",
    llm=ChatGoogleGenerativeAI(model="gemini-2.0-flash-lite-preview-02-05"),
    browser=browser,
)

async def main():
    await agent.run()

    input('Press Enter to close the browser...')
    await browser.close()

if __name__ == '__main__':
    asyncio.run(main())
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

from dotenv import load_dotenv
from browser_use import Agent
import asyncio

load_dotenv()

# Initialize the model
llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash-lite-preview-02-05", temperature=0.0)

# Define sensitive data
# The model will only see the keys (x_name, x_password) but never the actual values
sensitive_data = {'x_name': 'PearlHulbert', 'x_password': 'RoseTwitter@24'}

# Use the placeholder names in your task description
task = 'go to https://x.com/ and login with x_name and x_password then write a post about the meaning of life. Write it from the perspective of a cocky 22 year old salesbro'

# Pass the sensitive data to the agent
agent = Agent(task=task, llm=llm, sensitive_data=sensitive_data)

async def main():
    await agent.run()

if __name__ == '__main__':
    asyncio.run(main())
```
